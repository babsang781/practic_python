```python
# value 를 기준으로 데이터 합치기

df5 = pd.DataFrame({'key': ['K0','K2','K3','K4'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})


df6 = pd.DataFrame({'key': ['K0','K1','K2','K3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})
df5
```

```python
# 데이터를 기준으로 df 를 합침: merge()
# 기준 컬럼에서 공통된 내용만 합병, 공통된 컬럼이 여러개 있으면 지정해주어야함.

pd.merge(df5, df6, on = 'key')     # 기본 속성 how = inner
```

```python
pd.merge(df5, df6, on = 'key', how="outer")
```

```python
# how 속성 : left 와 right  합치는 기준을 왼쪽 df ,  오른쪽 df 
pd.merge(df5, df6, how="left")
```
