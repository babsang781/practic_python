```python
for j in range(1,10):
    url2 = 'https://movie.naver.com//movie/bi/mi/pointWriteFormList.naver?code=187320&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
    
    url2 = url2+str(j)  #리뷰 댓글 다음 페이지 링크 'page=' + str(j) 
    res = req.get(url2)  
    soup = bs(res.text, 'lxml')  # 변경된 페이지 soup 에 담기
    
    # 불필요한 정보(1): '관람객' (표시)삭제
    viewer = soup.select('span.ico_viewer')
    for i in viewer : 
        i.extract()  # extract() 함수는 soup 안에 담겨있는 정보를 추출(제거)
    
    # 불필요한 정보(2): '스포내용입니다.'' 삭제
    spo = soup.select("span.text_spo._text_spo")
    for i in spo:
        i.extract()
    
    review = soup.select('div.score_reple>p>span')
    for i in review:
        print(i.text.strip())
```
